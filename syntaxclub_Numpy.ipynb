 **Mathematical & Statistical Operations**

```python
import numpy as np

np.random.seed(42)

# Simulated dataset: 5 students, 4 subjects
scores = np.array([
    [85, 90, 78, 92],   # Student 0
    [70, 65, 88, 74],   # Student 1
    [95, 80, 72, 91],   # Student 2
    [60, 75, 85, 68],   # Student 3
    [88, 92, 79, 96],   # Student 4
])
subjects = ["Math", "Science", "English", "History"]
print("Score Matrix (5 students × 4 subjects):")
print(scores, "\n")
```

 ELEMENT-WISE ARITHMETIC

```python
print("═══ Element-wise Arithmetic ═══")
a = np.array([10, 20, 30, 40])
b = np.array([1,  2,  3,  4])

print(f"a + b   = {a + b}")
print(f"a - b   = {a - b}")
print(f"a * b   = {a * b}")
print(f"a / b   = {a / b}")
print(f"a ** 2  = {a ** 2}")
print(f"a % 3   = {a % 3}")
print(f"a // 3  = {a // 3}\n")
```

UNIVERSAL FUNCTIONS

```python
print("═══ Universal Functions ═══")
angles = np.array([0, np.pi / 6, np.pi / 4, np.pi / 3, np.pi / 2])

print(f"sqrt(a)     = {np.sqrt(a)}")
print(f"exp([1,2,3])= {np.exp([1, 2, 3])}")
print(f"log(a)      = {np.log(a)}")
print(f"sin(angles) = {np.round(np.sin(angles), 4)}")
print(f"cos(angles) = {np.round(np.cos(angles), 4)}")
print(f"abs([-3,4]) = {np.abs([-3, 4])}\n")
```

 AGGREGATION — FULL ARRAY

```python
print("═══ Full-Array Aggregation ═══")
print(f"Sum of all scores   : {np.sum(scores)}")
print(f"Min score           : {np.min(scores)}")
print(f"Max score           : {np.max(scores)}")
print(f"Mean score          : {np.mean(scores):.2f}")
print(f"Median score        : {np.median(scores):.2f}")
print(f"Std deviation       : {np.std(scores):.2f}")
print(f"Variance            : {np.var(scores):.2f}\n")
```

 AXIS-WISE OPERATIONS

```python
print("═══ Axis-wise Operations ═══")
print("  axis=0 → down rows (per subject)")
print("  axis=1 → across columns (per student)\n")

# Per subject (column-wise)
subject_means = np.mean(scores, axis=0)
print("Mean per subject :", dict(zip(subjects, np.round(subject_means, 2))))

subject_max = np.max(scores, axis=0)
print("Max per subject  :", dict(zip(subjects, subject_max)))

# Per student (row-wise)
student_means = np.mean(scores, axis=1)
print(f"\nMean per student : {np.round(student_means, 2)}")

student_totals = np.sum(scores, axis=1)
print(f"Total per student: {student_totals}")

# Cumulative sum per student
print(f"\nCumulative sum (axis=1):\n{np.cumsum(scores, axis=1)}\n")
```

SORTING & SEARCHING

```python
print("═══ Sorting & Searching ═══")

flat = np.array([42, 11, 88, 3, 55, 27])
print(f"Original       : {flat}")
print(f"Sorted         : {np.sort(flat)}")
print(f"Argsort (idx)  : {np.argsort(flat)}")

# Sort each student's scores
print(f"\nScores sorted per student (axis=1):\n{np.sort(scores, axis=1)}")

# Find indices where condition is true
print(f"\nIndices where flat > 30: {np.where(flat > 30)}")
print(f"Values where flat > 30: {flat[flat > 30]}")

# argmin / argmax
print(f"\nBest student (highest mean) : Student {np.argmax(student_means)}")
print(f"Hardest subject (lowest mean): {subjects[np.argmin(subject_means)]}")

```

FLATTEN vs RAVEL

```python
print("═══ Flatten vs Ravel ═══")

matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

flat_copy = matrix.flatten()    # returns a COPY
flat_view = matrix.ravel()      # returns a VIEW

print(f"flatten() : {flat_copy}  (copy)")
print(f"ravel()   : {flat_view}  (view)\n")

# Prove ravel is a view
flat_view[0] = 99
print(f"After ravel()[0] = 99 → matrix[0,0] = {matrix[0, 0]}")
matrix[0, 0] = 1  # reset
```

 STACKING & SPLITTING

```python
print("═══ Stacking & Splitting ═══")

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Stacking
print(f"hstack : {np.hstack([a, b])}")            # [1,2,3,4,5,6]
print(f"vstack :\n{np.vstack([a, b])}")            # 2×3
print(f"column_stack:\n{np.column_stack([a, b])}")  # 3×2
print(f"dstack shape: {np.dstack([a, b]).shape}")   # 1×3×2

# Concatenation
c = np.array([[1, 2], [3, 4]])
d = np.array([[5, 6], [7, 8]])
print(f"\nconcat axis=0:\n{np.concatenate([c, d], axis=0)}")  # 4×2
print(f"concat axis=1:\n{np.concatenate([c, d], axis=1)}")    # 2×4

# Splitting
big = np.arange(12).reshape(3, 4)
print(f"\nOriginal:\n{big}")
splits = np.hsplit(big, 2)   # split into 2 along columns
print(f"hsplit into 2:")
for i, s in enumerate(splits):
    print(f"  Part {i}: {s.shape}\n{s}")

vsplits = np.vsplit(big, 3)  # split into 3 along rows
print(f"vsplit into 3: {[s.shape for s in vsplits]}\n")
```

 BROADCASTING

```python
print("═══ Broadcasting ═══")
print("""
Broadcasting rules:
  1. If arrays differ in ndim, the smaller is padded with 1s on the left.
  2. Arrays with size 1 in a dimension act as if they had the size of
     the largest array in that dimension.
  3. If sizes disagree and neither is 1 → error.
""")

# Example 1: scalar + array
print("--- Scalar + Array ---")
arr = np.array([10, 20, 30])
result = arr + 5
print(f"[10, 20, 30] + 5 = {result}\n")

# Example 2: 1-D + 2-D (row broadcast)
print("--- Row Broadcast ---")
matrix = np.ones((3, 3)) * 10
row = np.array([1, 2, 3])
print(f"Matrix (3×3):\n{matrix}")
print(f"Row    (3,) : {row}")
print(f"Matrix + Row:\n{matrix + row}\n")

# Example 3: Column broadcast
print("--- Column Broadcast ---")
col = np.array([[100], [200], [300]])  # shape (3,1)
print(f"Col (3×1):\n{col}")
print(f"Matrix + Col:\n{matrix + col}\n")

# Example 4: Outer operation via broadcasting
print("--- Outer Product via Broadcasting ---")
x = np.array([1, 2, 3])          # shape (3,)
y = np.array([[10], [20], [30]])  # shape (3,1)
outer = x * y                    # shape (3,3)
print(f"x = {x}")
print(f"y =\n{y}")
print(f"x * y (outer product):\n{outer}\n")

# Example 5: Practical — Normalising columns (zero-mean, unit-std)
print("--- Practical: Column Normalisation ---")
data = np.array([[170, 65, 25],
                 [180, 80, 30],
                 [160, 55, 22],
                 [175, 70, 28]], dtype=float)
labels = ["Height", "Weight", "Age"]

col_mean = data.mean(axis=0)   # shape (4,) broadcasted over rows
col_std  = data.std(axis=0)

normalised = (data - col_mean) / col_std
print(f"Original:\n{data}")
print(f"Means : {col_mean}")
print(f"Stds  : {np.round(col_std, 2)}")
print(f"Normalised:\n{np.round(normalised, 2)}")

```

Save & Load Operations

```python
import numpy as np
import os
# Create an output directory for saved files
SAVE_DIR = os.path.join(os.path.dirname(os.path.abspath(__file__)), "saved_arrays")
os.makedirs(SAVE_DIR, exist_ok=True)
print(f"Save directory: {SAVE_DIR}\n")
```

```python
scores = np.array([
    [85, 90, 78, 92],
    [70, 65, 88, 74],
    [95, 80, 72, 91],
])
labels = np.array(["Alice", "Bob", "Charlie"])


# ─── 1. BINARY FORMAT — .npy (single array) ─────────────────

print("═══ 1. np.save / np.load (.npy) ═══")

filepath = os.path.join(SAVE_DIR, "scores.npy")
np.save(filepath, scores)
print(f"Saved scores to {filepath}")

loaded_scores = np.load(filepath)
print(f"Loaded scores:\n{loaded_scores}")
print(f"Arrays equal: {np.array_equal(scores, loaded_scores)}\n")


# ─── 2. MULTIPLE ARRAYS — .npz ──────────────────────────────

print("═══ 2. np.savez / np.savez_compressed (.npz) ═══")

# Uncompressed
filepath_npz = os.path.join(SAVE_DIR, "data_bundle.npz")
np.savez(filepath_npz, scores=scores, labels=labels)
print(f"Saved bundle to {filepath_npz}")

bundle = np.load(filepath_npz, allow_pickle=True)
print(f"Keys in bundle   : {list(bundle.keys())}")
print(f"bundle['scores'] :\n{bundle['scores']}")
print(f"bundle['labels'] : {bundle['labels']}\n")

# Compressed (smaller file size)
filepath_compressed = os.path.join(SAVE_DIR, "data_compressed.npz")
np.savez_compressed(filepath_compressed, scores=scores, labels=labels)

size_uncompressed = os.path.getsize(filepath_npz)
size_compressed   = os.path.getsize(filepath_compressed)
print(f"Uncompressed size: {size_uncompressed} bytes")
print(f"Compressed size  : {size_compressed} bytes")
print(f"Saved {size_uncompressed - size_compressed} bytes with compression\n")
```

```python

print("═══ 3. np.savetxt / np.loadtxt (.csv / .txt) ═══")

# Save as CSV (human-readable)
filepath_csv = os.path.join(SAVE_DIR, "scores.csv")
np.savetxt(filepath_csv, scores,
           delimiter=",",
           header="Math,Science,English,History",
           fmt="%d",
           comments="")
print(f"Saved CSV to {filepath_csv}")

# Print CSV contents
with open(filepath_csv, "r") as f:
    print(f"CSV contents:\n{f.read()}")

# Load from CSV
loaded_csv = np.loadtxt(filepath_csv, delimiter=",", skiprows=1)
print(f"Loaded from CSV:\n{loaded_csv}")
print(f"dtype: {loaded_csv.dtype}\n")


# ─── 4. HANDLING MISSING DATA — np.genfromtxt ───────────────

print("═══ 4. np.genfromtxt (missing data) ═══")

# Create a CSV with missing values
filepath_missing = os.path.join(SAVE_DIR, "incomplete_data.csv")
with open(filepath_missing, "w") as f:
    f.write("Height,Weight,Age\n")
    f.write("170,65,25\n")
    f.write("180,,30\n")        # missing weight
    f.write(",55,22\n")         # missing height
    f.write("175,70,\n")        # missing age

print(f"Created CSV with missing values at {filepath_missing}")
print("CSV contents:")
with open(filepath_missing, "r") as f:
    print(f.read())

# Load with genfromtxt — missing values become NaN
data = np.genfromtxt(filepath_missing, delimiter=",",
                     skip_header=1, filling_values=np.nan)
print(f"Loaded with genfromtxt (NaN for missing):\n{data}")
print(f"dtype: {data.dtype}\n")

# Replace NaN with column means
col_means = np.nanmean(data, axis=0)
print(f"Column means (ignoring NaN): {np.round(col_means, 2)}")

nan_mask = np.isnan(data)
for col in range(data.shape[1]):
    data[nan_mask[:, col], col] = col_means[col]

print(f"After filling NaN with column means:\n{np.round(data, 1)}")


# ─── 5. SUMMARY OF FORMATS ──────────────────────────────────

print("\n═══ Format Comparison ═══")
print(f"{'Format':<20} {'Extension':<10} {'Human Readable':<16} {'Multi-Array':<12} {'Compressed'}")
print("-" * 72)
print(f"{'np.save/load':<20} {'.npy':<10} {'No':<16} {'No':<12} {'No'}")
print(f"{'np.savez':<20} {'.npz':<10} {'No':<16} {'Yes':<12} {'No'}")
print(f"{'np.savez_compressed':<20} {'.npz':<10} {'No':<16} {'Yes':<12} {'Yes'}")
print(f"{'np.savetxt/loadtxt':<20} {'.csv/.txt':<10} {'Yes':<16} {'No':<12} {'No'}")
print(f"{'np.genfromtxt':<20} {'.csv/.txt':<10} {'Yes':<16} {'No':<12} {'No'}")

```

 Performance: NumPy vs Python Lists

```python
import numpy as np
import time
import sys

def benchmark(label, func, *args, runs=5):
    """Run a function multiple times and return avg time in ms."""
    times = []
    for _ in range(runs):
        start = time.perf_counter()
        result = func(*args)
        end = time.perf_counter()
        times.append((end - start) * 1000)
    avg = sum(times) / len(times)
    print(f"  {label:<35}: {avg:>8.3f} ms  (avg of {runs} runs)")
    return avg


SIZE = 1_000_000
print(f"═══ Performance Comparison: NumPy vs Python Lists ═══")
print(f"Array / List size: {SIZE:,}\n")
```

 CREATION

```python
print("─── 1. Creation ───")
t_list = benchmark("Python list (range)",
                    lambda: list(range(SIZE)))
t_np   = benchmark("NumPy arange",
                    lambda: np.arange(SIZE))
print(f"  → NumPy is {t_list / t_np:.1f}× faster\n")
```

ELEMENT-WISE MULTIPLICATION

```python

print("─── 2. Element-wise Multiply (× 2) ───")
py_list = list(range(SIZE))
np_arr  = np.arange(SIZE)

t_list = benchmark("Python list comprehension",
                    lambda: [x * 2 for x in py_list])
t_np   = benchmark("NumPy vectorised",
                    lambda: np_arr * 2)
print(f"  → NumPy is {t_list / t_np:.1f}× faster\n")
```

 SUM

```python

print("─── 3. Summation ───")
t_list = benchmark("Python sum()",
                    lambda: sum(py_list))
t_np   = benchmark("NumPy np.sum()",
                    lambda: np.sum(np_arr))
print(f"  → NumPy is {t_list / t_np:.1f}× faster\n"
```

FILTERING (BOOLEAN)

```python
print("─── 5. Boolean Filtering (> 500,000) ───")
t_list = benchmark("Python list comprehension",
                    lambda: [x for x in py_list if x > 500_000])
t_np   = benchmark("NumPy boolean indexing",
                    lambda: np_arr[np_arr > 500_000])
print(f"  → NumPy is {t_list / t_np:.1f}× faster\n")
```

MEMORY COMPARISON

```python
print("─── 6. Memory Usage ───")
py_mem     = sys.getsizeof(py_list) + sys.getsizeof(py_list[0]) * len(py_list)
np_mem     = np_arr.nbytes

print(f"  Python list : {py_mem / 1e6:.2f} MB")
print(f"  NumPy array : {np_mem / 1e6:.2f} MB")
print(f"  → NumPy uses {py_mem / np_mem:.1f}× less memory\n")
```
